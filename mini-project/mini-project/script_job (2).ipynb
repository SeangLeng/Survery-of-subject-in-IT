```python
from bs4 import BeautifulSoup
import pandas as pd
import re
import urllib.error
from urllib.request import urlopen
```

```python
# Define the URL of the main page with the list of cambodian jobs
main_page_url = 'https://www.cambojob.com/jobs/jobs_list/key/Developer/page/'

# Send an HTTP GET request to the main page and parse it with Beautiful Soup
main_page_response = urlopen(main_page_url)
main_page_soup = BeautifulSoup(main_page_response)
# print(main_page_soup)
main_page_soup.get_text()
```

```python


job_data_list = []

# Find all the job rows in the main page
job_rows = main_page_soup.find_all("li")

# Loop through each job row
for row in job_rows:
    job = row.find("h3").get_text()
    date= row.find("h5").find_all("span")
    d = [item.get_text() for item in date]
    
    # convert to string
    date_str = d[2]
    print(job.strip())
    print(date_str.strip())
    
    # Create a dictionary to store the data for this row
    job_data = {
        "Job": job.strip(),
        "Data": date_str.strip()
    }
    job_data_list.append(job_data)

# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(job_data_list)

# Save the DataFrame to a CSV file
df.to_csv("job_data.csv", index=False)

# Display the DataFrame
print(df)
```
